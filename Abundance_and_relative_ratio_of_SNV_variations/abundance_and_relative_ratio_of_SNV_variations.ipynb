```python
import pandas as pd

abd=pd.read_csv('D:\Workspace\PD_1\pan_cancer_analysis\Analysis_2024\Taxa_composion/data/adjust_abd_MEL.csv')
taxa=pd.read_csv('D:\Workspace\PD_1\pan_cancer_analysis\Analysis_2024\Taxa_composion/target_bact/data/unrepeat.csv')

df=pd.merge(taxa,abd,on='taxa')
df=df.drop(columns='taxa').rename(columns={'taxa_id':'taxa'})
data=df.T.reset_index()
data.columns=data.iloc[0,:]
abd=data.drop(index=0).rename(columns={'taxa':'run'})
snp_abd=pd.read_csv('D:\Workspace\PD_1\pan_cancer_analysis\Analysis_2024\Taxa_composion/target_bact/data/pivot_taxa_ko.csv')
snp_percent_abd=snp_abd.copy()

for i in range(len(snp_abd)):
    tmp=snp_percent_abd.iloc[i,1:]
    snp_percent_abd.iloc[i,1:]=snp_percent_abd.iloc[i,1:]/snp_percent_abd.iloc[i,1:].sum()*100
snp_percent_abd

meta=pd.read_csv('D:\Workspace\PD_1\pan_cancer_analysis\Analysis_2024\metadate_new.csv')
meta=meta[meta['cancer']=='MEL']
pro_list=meta[meta['cancer']=='MEL']['study_id'].unique().tolist()
taxa_list=pd.read_csv('D:\Workspace\PD_1\pan_cancer_analysis\Analysis_2024\Taxa_composion\target_bact\data\菌种list')
spe_list=taxa_list['specie'].tolist()
taxa_id_list=taxa_list['taxa'].tolist()
spe_list
```

```python
import collections
for specie,taxa in zip(spe_list,taxa_id_list):
    dic=collections.defaultdict(list)
    # print(taxa)
    print(f'-----------------------------{specie}------------------------------')
    # taxa='GCA_000154385.1'
    for pro in pro_list:
        meta_pro=meta[meta['study_id']==pro]
        R_list=meta_pro[meta_pro['response']=='R']['run']
        NR_list=meta_pro[meta_pro['response']=='NR']['run']
        abd_R=abd[abd['run'].isin(R_list)][taxa].mean()
        abd_NR=abd[abd['run'].isin(NR_list)][taxa].mean()
        snp_abd_R=snp_abd[snp_abd['run'].isin(R_list)][taxa].mean()
        snp_abd_NR=snp_abd[snp_abd['run'].isin(NR_list)][taxa].mean()

        snp_percent_abd_R=snp_percent_abd[snp_percent_abd['run'].isin(R_list)][taxa].mean()
        snp_percent_abd_NR=snp_percent_abd[snp_percent_abd['run'].isin(NR_list)][taxa].mean()
        # print(snp_abd[snp_abd['run'].isin(R_list)][taxa])
            
        # print(f'{pro}_____abd_R:{abd_R*100};abd_NR:{abd_NR*100}')
        # print(f'{pro}_____snp_abd_R:{snp_abd_R};SNP_abd_NR:{snp_abd_NR}')
        # print(f'{pro}_____snp_percent_abd_R:{snp_percent_abd_R};SNP_percent_abd_NR:{snp_percent_abd_NR}')

        Proportion_R=snp_percent_abd_R/(abd_R*100)
        Proportion_NR=snp_percent_abd_NR/(abd_NR*100)
        # print(f'{pro}_____R:{Proportion_R};NR:{Proportion_NR}')

        dic[f'{pro}_R'].extend([abd_R*100,snp_percent_abd_R,Proportion_R])
        dic[f'{pro}_NR'].extend([abd_NR*100,snp_percent_abd_NR,Proportion_NR])

    
    dic=pd.DataFrame(dic)
    dic.index=['abd','snp_abd','Proportion']
    dic.to_csv(f'D:\Workspace\PD_1\pan_cancer_analysis\Analysis_2024\Taxa_composion\target_bact\data\percent/{specie}.csv')
    print(dic) 
        
```
